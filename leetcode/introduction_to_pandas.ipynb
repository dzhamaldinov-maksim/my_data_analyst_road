3. Write a solution to select the name and age of the student with student_id = 101.

```python
import pandas as pd

def selectData(students: pd.DataFrame) -> pd.DataFrame:
    data = pd.DataFrame(students.loc[students['student_id'] == 101, ['name', 'age']])
    return data

"""
    Selects data from the students DataFrame for student with id 101
    and columns 'name' and 'age'.

    Parameters:
        students (pd.DataFrame): The DataFrame containing the student data.
        It is expected to have columns named 'student_id', 'name', and 'age'.

    Returns:
        pd.DataFrame: A new DataFrame containing the selected data.
    """
# 1. Loc: students.loc используется для выбора строк и столбцов фрейма данных students.

# 2. Условие фильтрации: students['student_id'] == 101 - это условие, которое выбирает только те строки фрейма данных students, где значение столбца student_id равно 101.

# 3. Столбцы для выбора: ['name', 'age'] - это список столбцов, которые необходимо выбрать из результата фильтрации.

# 4. Создание нового фрейма данных: Отфильтрованные строки и выбранные столбцы присваиваются новому фрейму данных data.
```

4. A company plans to provide its employees with a bonus.

Write a solution to create a new column name bonus that contains the doubled values of the salary column.

```python
import pandas as pd

def createBonusColumn(employees: pd.DataFrame) -> pd.DataFrame:
    employees['bonus'] = employees['salary']*2
    return employees
"""
    Создаёт в DataFrame employees новый столбец 'bonus'.
    Столбец bonus содержит в себе значения, в два раза большие значений
    столбца 'salary'.

    Parameters:
        employees (pd.DataFrame): DataFrame, содержащий данные о сотрудниках.
        Ожидается наличие столбца с названием 'salary'.

    Returns:
        pd.DataFrame: Изменённый DataFrame с добавленным столбцом 'bonus'.
"""
```

```python
import pandas as pd

def createBonusColumn(employees: pd.DataFrame) -> pd.DataFrame:
    return employees.assign(bonus = employees.salary * 2)
""" 
Creates a new column called 'bonus' in the employees DataFrame.
    The bonus column contains values that are twice the values 
    in the 'salary' column.

    Parameters:
        employees (pd.DataFrame): The DataFrame containing the employee data.
        It is expected to have a column named 'salary'.

    Returns:
        pd.DataFrame: The modified DataFrame with the added 'bonus' column.
P.S.
* employees.assign(bonus = employees.salary * 2) - создает новый столбец с именем bonus, содержащий значения, в два раза превышающие значения столбца salary. Метод assign() используется для добавления новых столбцов в DataFrame.

Этот код является более кратким способом создания столбца bonus.
"""
```
