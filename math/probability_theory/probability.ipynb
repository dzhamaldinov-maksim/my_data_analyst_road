1. **Независимые события**

Независимые события - это события, которые не влияют на вероятность друг друга. Другими словами, вероятность одного события не меняется в зависимости от того, произошло другое событие или нет.

**Математическое определение**\
Два события A и B являются независимыми, если и только если вероятность совместного события A и B равна произведению вероятностей A и B:

$$P(A \cap B) = P(A) \times P(B)$$

**Характеристики независимых событий**

* Вероятность совместного события A и B равна произведению вероятностей A и B.
* Вероятность условного события A при условии что произошло событие B равна вероятности A:


$$P(A | B) = P(A)$$


* Вероятность условного события B при условии что произошло событие A равна вероятности B:


$$P(B | A) = P(B)$$

**НЕЗАВИСИМОСТЬ СОБЫТИЙ**

Использование в аналитике данных:\
Независимость событий(умножение), широко используется в аналитике данных для:

* Расчета вероятности пересечения событий.
* Выявления взаимосвязей между переменными.
* Разработки предсказательных моделей.
* Принятия обоснованных решений на основе данных.

**ПРИМЕР**\
Вероятность события A равна 0,4, а вероятность события B равна 0,3. Вероятность совместного события A и B равна:

$$P(A \cap B) = 0,4 \times 0,3 = 0,12$$

Результат:\
Вероятность совместного события A и B равна 0,12.


**ПРИМЕР**\
Согласно правилу умножения вероятностей, вероятность вытянуть туза и черную карту равна:\
P(туз и черная карта) = P(туз) × P(черная карта) = 1/52 × 1/2 = 1/104 

Вероятность вытянуть туза при условии, что карта черная, равна:\
P(туз | черная карта) = P(туз и черная карта) / P(черная карта) = 1/104 / 1/2 = 1/52


**Применение в анализе данных**

Теория независимых событий широко используется в анализе данных для:

* Оценки вероятностей совместных событий
* Определения независимости между переменными
* Разработки моделей и алгоритмов для прогнозирования и принятия решений

2. **УСЛОВНАЯ ВЕРОЯТНОСТЬ. ПРАВИЛО УМНОЖЕНИЯ.**

**Условная вероятность**\
Условная вероятность события B, при условии, что уже произошло событие A, обозначается как P(B | A) и читается как "вероятность B при условии A".

$$P(B | A) = \frac{P(A \cap B)}{P(A)}$$

где:
* P(A ∩ B) - вероятность одновременного наступления событий A и B, также называемая совместной вероятностью.
* P(A) - вероятность наступления события A, называемая краевой вероятностью.

**Правило умножения**\
Правило умножения используется для расчета вероятности пересечения двух независимых событий, A и B.

$$P(A \cap B) = P(A) \times P(B)$$

где:
* P(A) - вероятность наступления события A.
* P(B) - вероятность наступления события B.

**УСЛОВНАЯ ВЕРОНЯТНОСТЬ**\
Допустим, у нас есть набор данных о 1000 клиентах интернет-магазина. Из них:

* 200 клиентов совершили покупку на сумму более 100 долларов.
* 600 клиентов являются мужчинами.
* 120 мужчин-клиентов совершили покупку на сумму более 100 долларов.

Мы хотим определить вероятность того, что клиент совершил покупку на сумму более 100 долларов, при условии, что он является мужчиной.

**1. Вероятность совершения покупки на сумму более 100 долларов (P(Покупка > 100)):**\
Эта вероятность рассчитывается путем деления количества клиентов, совершивших покупку на сумму более 100 долларов, на общее количество клиентов в наборе данных.

**2. Вероятность того, что клиент является мужчиной (P(Мужчина)):**\
Эта вероятность рассчитывается путем деления количества мужчин-клиентов на общее количество клиентов в наборе данных.

**3. Совместная вероятность того, что клиент совершил покупку на сумму более 100 долларов и является мужчиной (P(Покупка > 100 ∩ Мужчина)):**
Эта вероятность рассчитывается путем деления количества мужчин-клиентов, совершивших покупку на сумму более 100 долларов, на общее количество клиентов в наборе данных.

**4. Условная вероятность того, что клиент совершит покупку на сумму более 100 долларов при условии, что он является мужчиной (P(Покупка > 100 | Мужчина)):**

P(Покупка > 100 | Мужчина) = P(Покупка > 100 ∩ Мужчина) / P(Мужчина)

**Расчеты:**

1. P(Покупка > 100) = 200 / 1000 = 0,2
2. P(Мужчина) = 600 / 1000 = 0,6
3. P(Покупка > 100 ∩ Мужчина) = 120 / 1000 = 0,12
4. P(Покупка > 100 | Мужчина) = 0,12 / 0,6 = 0,2

**ТЕОРЕМА БАЙЕСА**

Теорема Байеса - это фундаментальный принцип теории вероятностей, который позволяет обновлять убеждения в свете новой информации.

$$
P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)}
$$


где:
* P(A | B) - вероятность события A при условии, что событие B уже произошло (задняя вероятность).
* P(B | A) - вероятность события B при условии, что событие A уже произошло (условная вероятность).
* P(A) - вероятность события A (априорная вероятность).
* P(B) - вероятность события B (краевая вероятность).

**Интерпретация**\
Теорема Байеса помогает нам понять, как новая информация влияет на наши убеждения:

* Знаменатель P(B) учитывает все возможные исходы, которые могли привести к событию B.
* Числитель P(B | A) * P(A) представляет собой вероятность наблюдения B при условии A, умноженную на априорную вероятность A. Априорная вероятность - это наша первоначальная вера в A до получения новой информации.
* Результат P(A | B) - это задняя вероятность A после учета новой информации B. Она представляет собой обновленное убеждение в A с учетом этой новой информации.

**ПРИМЕР**

Задача:\
Предположим, что в городе есть больница, в которой 0,1% всех пациентов больны редким заболеванием. Существует тест, который позволяет диагностировать это заболевание, но тест не всегда точен. Вероятность получения положительного результата теста у больного пациента составляет 99%, а у здорового пациента - 5%.

Некий пациент прошел этот тест и получил положительный результат. Какова вероятность того, что пациент действительно болен редким заболеванием?

Решение с использованием теоремы Байеса:

Пусть:
* A - событие, когда пациент болен редким заболеванием
* B - событие, когда пациент получает положительный результат теста

P(A|B) = (P(B|A) * P(A)) / P(B)

Вычисление вероятностей:
* P(B|A) = 0,99 (вероятность положительного результата теста у больного пациента)
* P(A) = 0,001 (вероятность того, что пациент болен редким заболеванием)
* P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)

P(B) = 0,99 * 0,001 + 0,05 * (1 - 0,001)\
P(B) = 0,00099 + 0,04995\
P(B) = 0,05094

P(A|B) = (0,99 * 0,001) / 0,05094\
P(A|B) ≈ 0,0194

Вероятность того, что пациент действительно болен редким заболеванием, составляет около 1,94%.
