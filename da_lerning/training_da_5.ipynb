```python
import pandas as pd
import numpy as np

frame = pd.DataFrame(np.arange(16).reshape((4,4)), index=['red','blue','yellow','white'], columns=['ball','pen','pencil','paper'])

    # np.sqrt() помогает нам найти квадратный корень всех значений

f = lambda x: x.max() - x.min()
    # лямбда функция, это любая функция которую мы определяем самостоятельно, она может как утроить значение, так и выполнить удвоение или любую другую операцию вложенную в нее. 
    # использовать лямбда функцию можно в любой части скрипта отдельно
    # выше мы определили лямбда функцию, которую мы создали сами
    # С помощью apply() новая функция применяется к Dataframe

frame.apply(f)
    # ball      12
    # pen       12
    # pencil    12
    # paper     12

frame.apply(f, axis=1)
    # red       3
    # blue      3
    # yellow    3
    # white     3

    # Можно также применить несколько функций одновременно, Но теперь результатом будет объект Dataframe, а не Series, в котором столько строк, сколько значений функция возвращает

def f(x):
    return pd.Series([x.min(), x.max()], index=['min','max'])
frame.apply(f, axis=1)
    # прописываем функцию и затем применяем ее к нашему объекту
    # так же мы можем указать ось к которой будет применена функция
```

```python
    # Два важных типа статистических вычислений — корреляция и вариантность. В pandas они представлены функциями corr() и cov(). Для их работы нужны два объекта Series.

import pandas as pd
import numpy as np

seq2 = pd.Series([3,4,3,4,5,4,3,2],['2006','2007','2008',
'2009','2010','2011','2012','2013'])
seq = pd.Series([1,2,3,4,4,3,2,1],['2006','2007','2008',
'2009','2010','2011','2012','2013'])
seq.corr(seq2)
# 0.7745966692414835
seq.cov(seq2)
# 0.8571428571428571

frame2 = pd.DataFrame([[1,4,3,6],[4,5,6,1],[3,3,1,5],[4,1,6,4]], index=['red','blue','yellow','white'], columns=['ball','pen','pencil','paper'])

frame2
frame2.corr()
    # расчитываем как данные в таблице коорелируют между собой

frame2.cov()
    # узнаем какоц разброс в данных

    # С помощью метода corwith() можно вычислить попарные корреляции между колонками и строками объекта Dataframe и Series или другим DataFrame().

    # frame2.corrwith(frame)
```
