```python
import sqlite3
import pandas as pd
import numpy as np

query = """CREATE TABLE sql02 (a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER);"""
    # создаем
con = sqlite3.connect(':memory:')
    #    - Эта строка создает новое подключение к базе данных SQLite, которая будет храниться в ОПЕРАТИВНОЙ ПАМЯТИ. Использование `':memory:'` означает, что база данных НЕ сохраняется на диске;
    # она будет существовать только во время работы программы и исчезнет, когда соединение будет закрыто или программа завершится.
con.execute(query)
    # - Эта строка выполняет SQL-запрос, который передан в переменной `query`. Этот запрос может быть любой допустимый SQL-код, например, создание таблицы, вставка данных, обновление и так далее.
con.commit()
    # - Эта строка фиксирует изменения, сделанные в базе данных. В SQLite все изменения базы данных, такие как вставка, обновление или удаление, не становятся постоянными до тех пор, пока не будет вызван метод `commit()`.
    # - Если не вызвать `commit()`, все изменения будут потеряны при закрытии соединения или завершении программы.


data = [('white','up',1,3), ('black','down',2,8), ('green','up',4,4), ('red','down',5,5)]

stmt = "INSERT INTO sql02 VALUES(?,?,?,?)"
    #  - Здесь создается строка SQL-запроса, которая предназначена для вставки данных в таблицу `sql02`. В этом запросе используются символы вопроса `(?, ?, ?, ?)`, которые представляют собой параметры для вставки. Это позволяет избежать SQL-инъекций и делает запрос более безопасным. На практике вы подставите значения вместо этих параметров.
con.executemany(stmt, data)
    # - Этот метод используется для выполнения SQL-запроса несколько раз с разными наборами данных. В данном случае `stmt` — это SQL-запрос, который мы определили ранее, а `data` предполагает собой последовательность кортежей, где каждый кортеж будет содержать значения, которые будут подставлены вместо параметров `?` в запросе. 
con.commit()


cursor = con.execute('select * from sql02')
cursor

rows = cursor.fetchall()
    # - Метод `fetchall()` используется для получения всех оставшихся строк результата последнего выполненного SQL-запроса. После выполнения запроса, когда вы хотите получить данные из базы данных, вы можете вызвать этот метод, чтобы получить все строки в виде списка.
rows

pd.DataFrame(rows, columns = list(zip(*cursor.description))[0])
    # **`cursor.description`**:
    # - Это свойство объекта `cursor`, которое возвращает информацию о колонках результата последнего выполненного SQL-запроса. Оно представляет собой список кортежей, где каждый кортеж содержит метаданные о конкретной колонке (например, имя поля, тип данных и т.д.)

    # **`zip(*cursor.description)`**:
    # - Оператор распаковки `*` используется для распаковки кортежей, содержащихся в `cursor.description`. Это позволяет сгруппировать информацию о колонках по соответствующим свойствам. Например, если `cursor.description` возвращает данные о колонках, такие как имя столбца и его тип, `zip(*cursor.description)` объединит эти данные в более удобный формат, где каждый элемент результирующего zip будет представлять собой данные для одного конкретного свойства (например, все имена столбцов).

    # **`list(zip(*cursor.description))`**:
    # - Приведение результата `zip` к списку позволяет нам получить доступ к каждому отдельному элементу по индексу. В результате получится список кортежей, где каждый кортеж будет содержать свойства на соответствующих позициях.
```
